```python
# 1 if anomali - 0 otherwise
df['label_bool'] = df['label'] == 'anomal'
df['label_bool'] = df['label_bool'].astype(int)
```

```python
import matplotlib.pyplot as plt

# regular is blue and anomal is red
colors = {0: 'blue', 1: 'red'}
df.plot.scatter('WAERS', 'BUKRS', c=df['label_bool'].map(colors), figsize=(15, 20))
```

```python
df[~df['WAERS'].str.contains('C')].plot.scatter('WAERS', 'BUKRS', c='label_bool', colormap='jet', figsize=(15, 20))
# TODO: check currencies: currencies not starting/ containing 'C' are anomalies
```
